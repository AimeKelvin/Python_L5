```python
#lambda 
hello = lambda x: "hello " + x

print(hello("kelvin"))

n = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, n))
print(even)
```

```python
#recursion
def countdown(n):
    if n == 0:
        print("We are Done Here!")
        return
    print(n)
    countdown(n-1)

print(countdown(20))
```

```python
#normal user defined function
def add(x,z):
    return x + z

print(add(5,10))
```
