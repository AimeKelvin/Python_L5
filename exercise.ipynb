```python

numbers = [3, 5, 3, 2, 6, 8, 1, 56]
product = 1
for num in numbers:
    if num % 2 != 0:  
        product *= num
print("Product of all odd numbers:", product)
```

```python
num = 10
count = 0

while count < 10:
    if num % 2 == 0:
        print(num)
        count += 1
    num += 1
```

```python

num = int(input("Enter a number: "))

print(f"\nMultiplication Table of {num}:")
for i in range(1, 13):
    print(f"{num} x {i} = {num * i}")
```
